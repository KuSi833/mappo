```python
import torch as th
class HDF5File():
    """
    PyTables has problems loading thousands of groups.
    H5Py is fine with it.
    """
    
    def __init__(self, path):
        self.path = path
        self.h5file = h5py.File(self.path, "r")
        pass
    
    def get_group_labels(self, test=True, _range=None):
        if _range is None:
            _range = (0, float("inf"))
        suffix = "_test" if test else ""
        timesteps = [ int(_k[1:]) for _k in self.h5file["learner_samples"+suffix].keys() if _range[0] <  int(_k[1:]) < _range[1] ]
        return ["T"+str(_t) for _t in sorted(timesteps)]
    
    def get_group(self, label, test=True):
        suffix = "_test" if test else ""
        return self.h5file["learner_samples"+suffix][label]
    
    def get_item(self, labels, item, test=True, scope="_transition", to_array=False, to_tensor=True):
        if not isinstance(labels, (tuple, list)):
            labels = [labels]
        from tables import open_file
        h5file = open_file(self.path, 
                   mode="r")
        suffix = "_test" if test else ""
        base = getattr(h5file.root, 
                       "learner_samples{}".format(suffix))
        if to_array:            
            return [ np.array(getattr(getattr(getattr(base, label), scope), item)) for label in labels ]
        elif to_tensor:
            return [ th.FloatTensor(getattr(getattr(getattr(base, label), scope), item)) for label in labels ]
        else:
            return [ getattr(getattr(getattr(base, label), scope), item) for label in labels ]
```
